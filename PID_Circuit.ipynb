## The model and dynamical equations 

The model of the PID controller is shown in the figure below:
![title](data/biomol_PID_scheme.png)

Our objecrtive is to control a simple process $\Phi(t)$, which describes the production of $X_{C}$ from $X_{1}$, with the subsequent decay/dilution of the product $X_{C}$:
\begin{equation}
\frac{dX_{C}}{dt} = \beta_{C}X_{1} - \gamma_{C}X_{C}
\end{equation}
<br>

The controller of the process is now described as follows:

\begin{align}
\frac{dZ_{1}}{dt} &= \mu Y - \gamma Z_{1}Z_{2} \\
\frac{dZ_{2}}{dt} &= \theta X_{C} - \gamma Z_{1}Z_{2} \\
\frac{dX_{1}}{dt} &= \beta_{I}Z_{1} + f_{P}(X_{C},Y) + D_{t}(X_{C},Y) - \gamma_{1}X_{1}
\end{align}

The first two equations essentialy describe the antithetic integral motif, while the third equation includes the proportional term $f_{P}(X_{C}, Y)$ and the derivative term $D_{t}(X_{C},Y)$, which are specified below. As in the paper by Chevalier et al., we will use the following forms of proportional controller:

\begin{align}
f_{P}(X_{C},Y) &= \beta_{P} Y \frac{\mu Y}{\mu Y + \theta X_{C}} \\
&= \frac{\beta_{P}}{2\mu} \left( \mu Y + \frac{\mu Y}{\mu Y + \theta X_{C}} \left[ \mu Y - \theta X_{C}  \right] \right)
\end{align}

Where the second form shows that this term essentially contains an offset proportional to $\mu Y$ and the error term $\mu Y - \theta X_{C}$, which vanishes if $X_{C}^{*}= \frac{\mu Y}{\theta}$.

We will also use the simplified form of derivative controller adopted from Ma *et al.*, which has been inspired by the bacterial chemotaxis, where bacteria needs to respond to the temporal differences in the concentration of the chemoattractants, such as essential amino acids. The controller is based on two molecules $A$ and $M$, so that $A$ is produced on the rate proportional to $M$ and actively degraded by the output $X_{C}$. On the other hand, $M$ is produced at the rate proportional to the reference signal $Y$ and actively degraded by $A$.

The dynamics of the two molecules can be described through the following set of equations: 

\begin{align}
\frac{dA}{dt} &= \beta_{A}M - \gamma_{A}X_{C} \frac{A}{K_{A} + A} - \gamma_{A_{0}}A \\
\frac{dM}{dt} &= \beta_{M}Y - \gamma_{M}A \frac{M}{K_{M} + M}
\end{align}

The circuit also requires that active degradation terms operate near saturation (which can be , for instance, achieved through strong binding afinity of the molecules $X_{C} \ (A)$ to $A \ (M)$). In that case, the approximate dynamics can be written as:

\begin{align}
\frac{dA}{dt} & \approx \beta_{A}M - \gamma_{A}X_{C} - \gamma_{A_{0}}A \\
\frac{dM}{dt} & \approx \beta_{M}Y - \gamma_{M}A
\end{align}

We take the time derivative of rate of change of $A$:

\begin{align}
\frac{d^{2}A}{dt^{2}} & \approx \beta_{A} \frac{dM}{dt} - \gamma_{A} \frac{dX_{C}}{dt} - \gamma_{A_{0}}\frac{dA}{dt}
\end{align}

We substitute for $\frac{dM}{dt}$ from the equation for the rate of change of $M$. After some rearranging, we obtain the following expression: 

\begin{align}
\frac{d^{2}A}{dt^{2}} + \gamma_{A_{0}} \frac{dA}{dt} + \beta_{A} \gamma_{M}A \approx -\gamma_{A} \frac{dX_{C}}{dt} + \beta_{A}\beta_{M}Y
\end{align}

If we assume that $\frac{d^{2}A}{dt^{2}} + \gamma_{A_{0}} \frac{dA}{dt} \ll \beta_{A} \gamma_{M}A$ (i.e. the rate of change of $\frac{dA}{dt}$ is low, for instance in the case of excessive $A$ being available), we can express the concentration of $A$ as:

\begin{align}
A = \frac{\beta_{M}Y}{\gamma_{M}} - \frac{\gamma_{A}}{\beta_{A}\gamma_{M}} \frac{dX_{C}}{dt}
\end{align}

This suggests that $A$ concentration is proportional to the negative time derivative of the concentration of the input $X_{C}$. Consequently, the derivative controller takes the form of $D_{t}(X_{C},Y) = \beta_{D}A$.

To summarize, the following set of equations describes the bacterial PID controller:

\begin{align}
\frac{dZ_{1}}{dt} &= \mu Y - \eta Z_{1}Z_{2} \\
\frac{dZ_{2}}{dt} &= \theta X_{C} - \eta Z_{1}Z_{2} \\ \\
\frac{dX_{C}}{dt} &= \beta_{C}X_{1} - \gamma{C}X_{C} \\
\frac{dA}{dt} &= \beta_{A}M - \gamma_{A}X_{C} - \gamma_{A_{0}}A \\
\frac{dM}{dt} &= \beta_{M}Y - \gamma_{M}A \\
\frac{dX_{1}}{dt} &= \beta_{I}Z_{1} + \beta_{P} Y \frac{\mu Y}{\mu Y + \theta X_{C}} + \beta_{D}A - \gamma_{1}X_{1} \\
\end{align}

In the following we might refer to the above set of the equations as

\begin{align}
\frac{d \textbf{x}}{dt} = f(\textbf{x}, u, \theta)
\end{align}
Where $\textbf{x} = (Z_{1}, Z_{2}, X_{C}, A, M, X_{1})$ is the vector of states, $u = X_{1}$ is the input, and $\theta$ is the vector of parameters.

To run the simulation, we use the following set of parameters:
* $\beta_{I} = 0.06 \ min^{-1}$
* $\beta_{P} = 0.15 \ min^{-1}$
* $\beta_{D} = 0.28 \ min^{-1}$
* $\mu = 1 \ min^{-1}$
* $\eta = 0.01 \ nM^{-1} \ min^{-1}$
* $\theta = 0.3 \ min^{-1}$
* $\gamma_{1} = 0.1 \ min^{-1}$
* $\beta_{C} = 0.1 \ min^{-1}$
* $\gamma_{C} = 0.1 \ nM$
* $\gamma_{A_{0}} = 0.1 \ min^{-1}$
* $\beta_{A} = 1.5 \ min^{-1}$
* $\gamma_{A} = 1.5 \ min^{-1}$
* $\gamma_{M} = 1.5 \ min^{-1}$
* $K_{A} = 1 \ nM$
* $K_{M} = 1 \ nM$
* $\beta_{M} = 0.4167 \ min^{-1}$

And now let's delve into the coding! 


```python
# Use odeint() for the small problems where generic LSODA algorithm by Petzold and Hindmarsh is used
# Define the right hand side of the function
def pid_controller_rhs(states, t, 
                       beta_I, beta_P, beta_D, 
                       mu, eta, theta, 
                       gamma_1, beta_C, gamma_C, gamma_A0, 
                       beta_A, gamma_A, gamma_M, 
                       K_A, K_M, beta_M, 
                       Y):
    """
    The right hand side for the PID controller. Returns dZ1/dt, dZ2/dt, dXC/dt, dA/dt, dM/dt, dX1/dt.
    ------
    Inputs: 
    - states(list): list of concentrations of chemical species for which we aim to determine the time evolution
    - t(float): a variable describing the time evolution
    - pars(list): a vector of parameters that enter the ODEs, including the reference Y,
    which is the last parameter of the array 
    ------
    Outputs:
    - pid_rhs(np.array): an array of the right hand sides of the system of the ODE
    """
    
    # Unpack the state variables
    Z1, Z2, Xc, A, M, X1 = states
    
    # Compute dZ1/dt
    dZ1_dt = mu*Y - eta*Z1*Z2
    
    # Compute dZ2/dt
    dZ2_dt = theta*Xc - eta*Z1*Z2
    
    # Compute dXc/dt
    dXc_dt = beta_C*X1 - gamma_C*Xc
    
    # Compute dA/dt
    dA_dt = beta_A*M - gamma_A*Xc - gamma_A0*A
    
    # Compute dM/dt
    dM_dt = beta_M*Y - gamma_M*A
    
    # Compute dX1/dt
    dX1_dt = beta_I*Z1 - gamma_1*X1 + beta_P*Y*mu*Y/(mu*Y + theta*Xc) + beta_D*A
    
    # Return the np.array() of RHS of the ODEs
    pid_rhs = np.array([dZ1_dt, dZ2_dt, dXc_dt, dA_dt, dM_dt, dX1_dt])
    
    return pid_rhs
```

```python
### Run the simulation
# Number of the points for which we want the solution
n = 1000

# Simulation length
t_f = 4000 # min
# Time points at which we want to have the solution calculated
t = np.linspace(0, t_f, n)

# Initial values of states
Z1 = 0.0
Z2 = 0.0
Xc = 0.0
A = 0.0
M = 0.0
X1 = 0.0

# Pack the initial values of states
states_init = np.array([Z1, Z2, Xc, A, M, X1])

# The reference to be tracked
Y_0 = 600 # nM

# The vector representing the reference Y with the same number of points as time vector,
# this is for plotting
Y_vec = Y_0*np.ones(n)

# Define the parameters
beta_I = 0.06
beta_P = 0.15
beta_D = 0.28

mu = 1
eta = 0.01
theta = 0.3
gamma_1 = 0.1
beta_C = 0.1
gamma_C = 0.1
gamma_A0 = 0.1
beta_A = 1.5
gamma_A = 1.5
gamma_M =1.5
K_A = 1
K_M = 1
beta_M = 0.4167

# Pack the parameters into a tuple
args = (beta_I, beta_P, beta_D, mu, eta, theta, 
     gamma_1, beta_C, gamma_C, gamma_A0, 
     beta_A, gamma_A, gamma_M, K_A, K_M, beta_M, Y_0)

# Integrate the ODEs
states_out = scipy.integrate.odeint(pid_controller_rhs, states_init, t, args=args)
```

```python
# Pluck out y and z
Z1, Z2, Xc, A, M, X1 = states_out.transpose()

# Plot the results
fig1, ax1 = plt.subplots(1,1, figsize=(9,6))
ax1.plot(t, Xc, linewidth=2, color='red', label='Xc')
ax1.plot(t,Y_vec, linewidth=2, color='seagreen', label='Y')
ax1.plot(t,X1, linewidth=2, color='royalblue', label='X1')

ax1.legend(fontsize=15)
ax1.set_xlabel('Time (min)', fontsize=20)
ax1.set_ylabel('Concentration (nM)', fontsize=20)
ax1.set_title('Time evolution of the PID controller', fontsize=25)

plt.show(fig1)
```

```python
def Y_pulse(t, t_0, tau, Y_mag, off=0.0):
    """
    Returns Y value for a pulse beginning at t_0 and 
    ending at t = t_0 + tau with magnitude Y0, and an offset 'off'
    """
    return off + np.logical_and(t >= t_0, t <= (t_0 + tau)) * Y_mag
            

# Plot the pulse for some values
fig1, ax1 = plt.subplots(1,1, figsize=(9,6))
ax1.plot(t, Y_pulse(t, 100.0, 200.0, 300.0, 300.0), linewidth=2, color='red', label='Y(t)')

ax1.set_xlabel('Time (min)', fontsize=20)
ax1.set_ylabel('Concentration (nM)', fontsize=20)
ax1.set_title('Input pulse', fontsize=25)
ax1.legend(fontsize=15)
plt.show(fig1)
```

```python
def pid_controller_input_fun(states, t,
                             beta_I, beta_P, beta_D, 
                             mu, eta, theta, 
                             gamma_1, beta_C, gamma_C, gamma_A0, 
                             beta_A, gamma_A, gamma_M, 
                             K_A, K_M, beta_M, 
                             Y_fun, Y_args):
    """
    Right hand side for the PID controller with the input waveform.  
    Returns dZ1/dt, dZ2/dt, dXC/dt, dA/dt, dM/dt, dX1/dt.
    ------
    Inputs: 
    - states(list): list of concentrations of chemical species for which we aim to determine the time evolution
    - t(float): a variable describing the time evolution
    - pars(list): a vector of parameters that enter the ODEs, EXCLUDING the reference Y
    
    Y_fun is a function of the form Y_fun(t, *Y_args), so Y_args is a tuple
    containing the arguments to pass to Y_fun.
    ------
    Outputs:
    - pid_rhs(np.array): an array of the right hand sides of the system of the ODE
    """
    
    # Compute the reference Y
    Y = Y_fun(t, *Y_args)
        
    # Return the PID controller RHS with the given value of Y
    return pid_controller_rhs(states, t,
                              beta_I, beta_P, beta_D, 
                              mu, eta, theta, 
                              gamma_1, beta_C, gamma_C, gamma_A0, 
                              beta_A, gamma_A, gamma_M, 
                              K_A, K_M, beta_M, 
                              Y)
```

```python
# Solve the ODEs for the pulse
tp_init = 500.0
tp_duration = 2000.0
Y_mag = 300.0
Y_off = 0.01
Y_args = (tp_init, tp_duration, Y_mag, Y_off)

# Pack the parameters into a tuple 
args = (beta_I, beta_P, beta_D, mu, eta, theta, 
     gamma_1, beta_C, gamma_C, gamma_A0, 
     beta_A, gamma_A, gamma_M, K_A, K_M, beta_M, Y_pulse, Y_args)

# Initial values of states
Z1 = 0.00
Z2 = 0.00
Xc = 0.00
A = 0.00
M = 0.00
X1 = 0.00

# Initial conditions
states_init = np.array([Z1, Z2, Xc, A, M, X1])

# Integrate the ODEs
states_out = scipy.integrate.odeint(pid_controller_input_fun, states_init, t, args=args)
```

```python
# Pluck out y and z
Z1, Z2, Xc, A, M, X1 = states_out.transpose()

### Plotting 
# Plot the results for the main species
fig1, ax1 = plt.subplots(1,1, figsize=(9,6))
ax1.plot(t, Xc, linewidth=3, color='seagreen', label='Output Xc(t)')
ax1.plot(t, Y_pulse(t, tp_init, tp_duration, Y_mag, Y_off), linewidth=3, color='red', label=' Reference Y(t)')
ax1.plot(t, Y_pulse(t, tp_init, tp_duration, Y_mag, Y_off)*mu/theta, linestyle='--', linewidth=3, alpha=0.8, color='black',
         label=('Steady-state Output'))
ax1.legend(fontsize=15)
ax1.set_xlabel('Time (min)', fontsize=20)
ax1.set_ylabel('Concentration (nM)', fontsize=20)
ax1.set_title('Time evolution of PID controller', fontsize=25)

# Plot the results for the other species
fig2, ax2 = plt.subplots(1,1, figsize=(9,6))
ax2.plot(t, Z1, linewidth=2, color='orange', label='Z1(t)')
ax2.plot(t, X1, linewidth=2, color='royalblue', label='X1(t)')
ax2.plot(t, Y_pulse(t, tp_init, tp_duration, Y_mag), linewidth=2, color='red', label=' Reference Y(t)')

ax2.legend(fontsize=15)
ax2.set_xlabel('Time (min)', fontsize=20)
ax2.set_ylabel('Concentration (nM)', fontsize=20)
ax2.set_title('Time evolution of PID controller', fontsize=25)

plt.show(fig2)
```
